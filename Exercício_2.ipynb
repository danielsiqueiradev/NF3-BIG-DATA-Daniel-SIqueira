```python
#Tabela base importada do .csv

import pandas as pd

df = pd.read_csv('https://raw.githubusercontent.com/danielsiqueiradev/NF3-BIG-DATA-Daniel-SIqueira/main/cursos-prouni.csv', encoding = 'latin-1', sep = ';')

df
```

```python
#Exercício 2.a)
#a. Efetuar a limpeza das colunas de notas: onde tiver NaN (Not a Number), substituir por 0,0.


df = df.apply(lambda x: x.fillna('0.0'))
df
```

```python
#Exercício 2.b)
#b. Agrupe os dados pelo grau (Bacharelado, Licenciatura, etc).


import pandas as pd

grupo_grau = df.groupby('grau')

for grau, dados in grupo_grau:
    print(f"Grau: {grau}")
    print(dados)
    print()
```

```python
#Exercício 2.c)
#c. Agrupe os dados pelos cursos de Matemática, Medicina e Pedagogia.

cursos_selecionados = ['Matemática', 'Medicina', 'Pedagogia']
filtro = df['curso_busca'].isin(cursos_selecionados)

grupo_cursos = df[filtro].groupby('curso_busca')

for curso, dados in grupo_cursos:
    print(f"Curso: {curso}")
    print(dados)
    print()
```

```python
#Exercício 2.d)
#d. Agrupe os dados por Estado e obtenha a média de notas de corte por Estado.

# grupo_estados = df.groupby('uf_busca')
# notas = grupo_estados['nota_integral_ampla']
# notas_corrigidas = notas.mean()
# notas_corrigidas.describe()

# # filtro = dados[dados['Year'] == 1985]
# # filtro_bebidas = filtro['Beverage Types']
# # filtro_bebidas

filtro = df[['nota_integral_ampla', 'nota_integral_cotas', 'nota_parcial_ampla', 'nota_parcial_cotas']].replace(0, float('inf'))

media_cortes_estado = filtro.groupby(df['uf_busca']).min().mean(axis=1)
media_cortes_estado


```

```python
#Exercício 2.e)
#e. Agrupe os dados pelos cursos Tecnológicos.

filtro_tecnologicos = df['grau'].str.contains('Tecnológico', case=False)
grupo_tecnologicos = df[filtro_tecnologicos].groupby('grau')

for curso, dados in grupo_tecnologicos:
    print(f"Curso: {curso}")
    print(dados)
    print()
```

```python
#Exercício 2.f)
#f. Elimine a coluna “cidade_filtro” do dataframe.

df = df.drop('cidade_filtro', axis=1)

df
```

```python
#Exercício 2.g)
#g. Apresente a média das mensalidades dos cursos de Medicina.

# df['mensalidade'] = df['mensalidade'].str.replace(',', '.').astype(float)
# df['mensalidade'].info()


grupo_medicina = df[df['curso_busca'] == 'Medicina']
mensalidade = grupo_medicina['mensalidade']
media_mensalidade = mensalidade.mean()

print('A média da mensalidade nos cursos de Medicina é,',media_mensalidade)


```

```python
#Exercício 2.h)
#h. Média das notas de corte dos cursos de tempo integral.

# df['nota_integral_ampla'] = df['nota_integral_ampla'].str.replace(',', '.').astype(float)
# df['nota_integral_ampla'].info()


grupo = df[df['turno'] == 'Integral']
nota = grupo['nota_integral_ampla']
media_notas = nota.mean()

print('A média da mensalidade nos cursos em tempo Integral é,',media_notas)
```

```python
#Exercício 2.i)
#i. Estatística Descritiva das Notas Integral Ampla dos cursos de Bacharelado.

grupo = df[df['turno'] == 'Integral']
nota = grupo[df['grau'] == 'Bacharelado']

nota.describe()

```

```python
#Exercício 2.j)
#j. Gráfico comparativo entre o grau dos cursos (Bacharelado, Licenciatura, Tecnologia, etc) pelas Notas Integral de Cotas.

# df['nota_integral_cotas'] = df['nota_integral_cotas'].str.replace(',', '.').astype(float)
# df['nota_integral_cotas'].info()

group = df.groupby('grau')['nota_integral_cotas'].std()

group.plot.bar(x='grau', y = 'nota_integral_ampla')


```
